**Clean up data**

USA races, 50km or 50mi, 2020

step 1: show 50km or 50mi

```python
df[df['Event distance/length'] == '50km']
```

```python
# combine 50km/50mi with isin()
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
# events in USA
df[df['Event name'].str.extract(r'\((.*?)\)')[0] == 'USA']
```

```python
# combine all filters
# 50km/50mi & 2020 & USA
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.extract(r'\((.*?)\)')[0] == 'USA')]
df2 = df2.copy()
```

```python
# remove USA from event name
# Use .loc to avoid SettingWithCopyWarning
df2.loc[:, 'Event name'] = df2['Event name'].str.split('(').str.get(0)
df2.head()
```

```python
# clean up athlete age
df2.loc[:,'athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# remove h from Athelete performance
df2.loc[:, 'Athlete performance'] = df2['Athlete performance'].str.replace('h', '')
df2.head()
```

```python
# drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
df2.drop(columns=['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], inplace=True)
df2.head()
```

Clean up null values

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
df2.isna().sum()
```

Check for duplicates

```python
df2[df2.duplicated() == 1]
```

Reset index

```python
df2.reset_index(drop=True)
```

Fix types

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int) 
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float) 
```

```python
df2.dtypes
```

Rename columns

```python
df2 = df2.rename(columns={'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id',
    })
```

```python
df2.head()
```

Reorder columns

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
df3.head()
```

**Questions I want to find out from data**

```python
# Difference in speed for the 50km/50mi male to female
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what age groups are best in the 50mi race(20+ races min) (show 15)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

```python
# Seasons for the data -> slower in summer than in winter

# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2

# split between two decimals
df3.head()
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)  # european format dd.mm.yyyy
df3.head()
```

```python
df3['race_season'] = df3['race_month'].map({3: 'spring', 4: 'spring', 5: 'spring', 6: 'summer', 7: 'summer', 8: 'summer', 9: 'fall', 10: 'fall', 11: 'fall', 12: 'winter', 1: 'winter', 2: 'winter'})
```

```python
df3.sample(5)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False) 
```

```python
# 50 miler only
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False) 
```
